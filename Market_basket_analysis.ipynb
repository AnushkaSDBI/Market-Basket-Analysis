```python

txn_list = []

for i in range(0,df.shape[0]):
  row_list = []
  for j in range(0,df.shape[1]):
    if df_list[i][j]!="":
      row_list.append(df_list[i][j])
  txn_list.append(row_list)
```

```python
rules = apriori(transactions = txn_list,min_support  = .0002,min_conf=.5,min_lift=3,min_length = 1,max_length = 2 )
```

```python
#new dataset
df_rules = pd.DataFrame()
```

```python
rules = []
support = []
lhs = []
rhs = []
conf = []
lift = []
for i in range(0,len(list_rules)):
  for j in range(0,len(list_rules[i])):
    if j == 0:
      rules.append(list_rules[i][j])   
    if j == 1:
      support.append(list_rules[i][j])
    if j == 2:
      str_list = str(list_rules[i][j])
      details = str_list.split(",")
      lhs.append(details[0])
      rhs.append(details[1])
      conf.append(details[2])
      lift.append(details[3])
```

```python
rules
```

 Making the new dataframe

```python
df_rules["rules"]=rules
df_rules["support"]= support 
df_rules["lhs"]=lhs
df_rules["rhs"]=rhs
df_rules["conf"]=conf
df_rules["lift"]=lift
```
